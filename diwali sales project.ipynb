# why do we use inplace=True


```python
test=[["madhu",15],['raghu',14],['rani',16]]

test1=pd.DataFrame(test,columns=['name','age'])
test1
```

```python
test=[["madhu",15],['raghu',14],['rani',16],['keshav',]]

test1=pd.DataFrame(test,columns=['name','age'])
test1
```
